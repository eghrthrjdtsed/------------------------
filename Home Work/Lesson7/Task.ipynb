Распишите, какие метрики и как вы будете анализировать</h4>

<h4 style='color: blue;'>OZON.ru</h4>
<em>(Количество новых пользователей, время на сайте, количество кликов, глубина просмотра, средний чек заказа)</em>

<em><b>Думаю, что эти метрики покажут суть того чего мы хотим и понять в каком месте нам необходимо внести те или иные изменения.</b></em>
1. Кол-во новых пользователей для нас важно т.к это отражает успех и правильное направление нашей работы.

2. Время на сайте отражает то, что наш контент качественный и интересный.

3. Количество кликов говорит нам, что пользователь был заинтересован в продукте.

4. По глубине просмотра можем понять, какое количество страниц просмотрел пользователь во время одного сеанса.

5. Ну и средний чек заказа нам важен, он отражает какие отдельные магазины приносят больше выручки.

Ответ: Возьмем метрики 1 и 5 посчитаем на сколько дней можем запустить а/в-тест, далее посчитаем критерий z или хи-квадрат. И на основе полученных данных проводим анализ и принимаем дальнейшее решение. Всех замотивированных просим соблюдать режим и сроки проведения теста и анализа.
